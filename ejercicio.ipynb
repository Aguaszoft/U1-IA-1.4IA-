```python
# prompt: Ejercicio 4: Búsqueda en Profundidad (DFS - Depth First Search)
# Enunciado: Escribe una función en Python que implemente la búsqueda en profundidad (DFS) en un grafo representado como un diccionario. La función debe visualizar el proceso de exploración, destacando los nodos visitados y el camino actual. Si encuentra el nodo objetivo, debe detenerse y resaltar el recorrido.
# • Grafo de ejemplo:
# grafo = {
# 'A': ['B', 'C'],
# 'B': ['D', 'E'],
# 'C': ['F'],
# 'D': [],
# 'E': ['F'],
# 'F': []
# }
# • Inicio: 'A'
# • Objetivo: 'F'

def dfs_visual(graph, start_node, target_node):
  """
  Realiza una búsqueda en profundidad (DFS) en un grafo y visualiza el proceso.

  Args:
    graph: El grafo representado como un diccionario.
    start_node: El nodo de inicio.
    target_node: El nodo objetivo.
  """

  visited = set()  # Conjunto para llevar un registro de los nodos visitados
  stack = [(start_node, [start_node])]  # Pila para almacenar los nodos a explorar (nodo, camino)

  print("Iniciando DFS desde el nodo:", start_node)

  while stack:
    (node, path) = stack.pop()

    if node not in visited:
      print("Visitando nodo:", node)
      visited.add(node)

      if node == target_node:
        print("¡Objetivo encontrado!")
        print("Camino:", path)
        return

      neighbors = graph[node]
      for neighbor in reversed(neighbors):  # Se invierte para explorar en orden DFS
        if neighbor not in visited:
          stack.append((neighbor, path + [neighbor]))
          print("Agregando a la pila:", neighbor)
          print("Camino actual:", path + [neighbor])
  print("Objetivo no encontrado.")


# Ejemplo de uso
grafo = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

dfs_visual(grafo, 'A', 'F')
```
